```python
#Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusts the configuration of the plots we will create
```

```python
#Read in the data

df = pd.read_csv(r'/Users/angela.kyea/Downloads/bestsellers with categories.csv')
```

```python
#1. 

#Null Hypothesis_A: The result of the correlation analysis between Price & User Rating proves that there is not 
#a positive, high correlation between Price and User Rating. 

#Alternative Hypothesis_A: The result of the correlation analysis between Price & User Rating proves that there 
#exists a positive, high correlation between Price and User Rating. 


#Plotting price against user rating to determine if there's correlation

sns.regplot(x='User Rating', y='Price', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
plt.title("User Rating vs Price")
plt.xlabel("User Rating")
plt.ylabel("Price ($)") 
```

```python
#2. 

#Null Hypothesis_B: The result of the correlation analysis between Price & Reviews proves that there is not 
#a positive, high correlation between Price and Reviews. 

#Alternative Hypothesis_B: The result of the correlation analysis between Price & Reviews proves that there 
#exists a positive, high correlation between Price and Reviews.


#Plotting price against reviews to determine if there's correlation

sns.regplot(x='Reviews', y='Price', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
plt.title("Reviews vs Price")
plt.xlabel("Reviews")
plt.ylabel("Price ($)") 
```

```python
#Using the Pearson Correlation Coefficient Method

df.corr(method='pearson')
```

```python
#According to the table above:

#The correlation coefficient of price and user rating is -0.133086.
#In conclusion, there exists a negative, negligible correlation between price and user rating.
#Therefore, I do not reject Null Hypothesis_A. 
#This means that a high user rating on a book in this dataset does not mean that the book will have a higher price.

#The correlation coefficient of price and reviews is -0.109182.
#In conclusion, there exists a negative, negligible correlation between price and reviews.
#Therefore, I do not reject Null Hypothesis_B. 
#This means that a high number of reviews on a book in this dataset does not mean that the book will have a higher price.
```
